```python
import pandas as pd
from random import randrange
from datetime import timedelta
from datetime import datetime
import numpy as np
RNG = np.random.default_rng()
```

## KOLEJNOŚĆ DODAWANIA DANYCH:
#### COUNTRIES -> LEAGUES -> TEAMS -> PLAYERS -> STADIUMS -> MATCHES -> TEAM_PLAYS_IN_LEAGUE -> TEAM_PLAYS_IN_MATCH


```python
df1 = pd.read_csv('source_csv/2022-2023 Football Player Stats.csv', sep=';', encoding_errors='ignore')
df1 = df1[df1['Player'].str.contains('\?')==False]
df1
```

```python
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return str(start + timedelta(seconds=random_second))[:10]

d1 = datetime.strptime('1/1/1980 1:30 PM', '%m/%d/%Y %I:%M %p')
d2 = datetime.strptime('1/1/2010 4:50 AM', '%m/%d/%Y %I:%M %p')
```

```python
players = df1[['Player', 'Rk']].sample(n=500)
players[['name', 'surname']] = players['Player'].str.split(' ', 1, expand=True)
players.rename(columns={'Rk':'player_id'}, inplace=True)
players.drop(columns='Player', inplace=True)
players['birth_date'] = [random_date(d1, d2) for _ in range(players.shape[0])]
players['height'] = [int(RNG.normal(180, 15)) for _ in range(players.shape[0])]
players['weight'] = [int(RNG.normal(85, 15)) for _ in range(players.shape[0])]
```

```python
df2 = pd.read_csv('source_csv/countries of the world.csv', sep=',')
countries = df2[['Country']]
countries.rename(columns={'Country':'name'}, inplace=True)
countries['country_id'] = countries.index
```

```python
players['birth_country_id'] = [int(countries['country_id'].sample(n=1)) for _ in range(players.shape[0])]
```

```python
df3 = pd.read_csv('source_csv/Football teams.csv', sep=',')
leagues = pd.DataFrame()
leagues['name'] = pd.unique(df3['Tournament'])
leagues['league_id'] = leagues.index
```

```python
teams = df3[['Team', 'Tournament']]
teams.rename(columns={'Team':'name'}, inplace=True)
teams['team_id'] = teams.index
teams['acronym'] = teams['name'].str.slice(0, 3).str.upper()
teams.drop(columns='Tournament', inplace=True)
teams['country_id'] = [int(countries['country_id'].sample(n=1)) for _ in range(teams.shape[0])]
```

```python
players['team_id'] = [int(teams['team_id'].sample(n=1)) for _ in range(players.shape[0])]
```

```python
df4 = pd.read_csv('source_csv\stadiums.csv', sep=',')
stadiums = df4[['Name']].sample(n=50)
stadiums['stadium_id'] = stadiums.index
stadiums.rename(columns={'Name':'name'}, inplace=True)
stadiums['country_id'] = [int(countries['country_id'].sample(n=1)) for _ in range(stadiums.shape[0])]
```

```python
team_plays_in_league = pd.DataFrame()
team_plays_in_league['team_id'] = teams['team_id']
team_plays_in_league['league_id'] = [int(leagues['league_id'].sample(n=1)) for _ in range(team_plays_in_league.shape[0])]
team_plays_in_league
```

```python
matches = pd.DataFrame()
d3 = datetime.strptime('1/1/2020 1:30 PM', '%m/%d/%Y %I:%M %p')
d4 = datetime.strptime('11/29/2022 4:50 AM', '%m/%d/%Y %I:%M %p')
matches['match_date'] = [random_date(d1, d2) for _ in range(1000)]
matches['match_id'] = matches.index
matches['stadium_id'] = [int(stadiums['stadium_id'].sample(n=1)) for _ in range(matches.shape[0])]
matches['league_id'] = [int(leagues['league_id'].sample(n=1)) for _ in range(matches.shape[0])]
matches
```

```python
team_plays_in_match = pd.DataFrame(columns = ['team_id', 'match_id', 'goal_amount'])
for match_id in matches['match_id']:
    for team_id in teams['team_id'].sample(n=2).tolist():
        team_plays_in_match = team_plays_in_match.append(pd.DataFrame({'team_id': team_id,
                                                                       'match_id': [match_id],
                                                                       'goal_amount': [RNG.integers(0, 5)]}), ignore_index = True)
team_plays_in_match
```

```python
players.to_csv('output_csv/players.csv', index=False)
countries.to_csv('output_csv/countries.csv', index=False)
teams.to_csv('output_csv/teams.csv', index=False)
leagues.to_csv('output_csv/leagues.csv', index=False)
stadiums.to_csv('output_csv/stadiums.csv', index=False)
matches.to_csv('output_csv/matches.csv', index=False)
team_plays_in_league.to_csv('output_csv/team_plays_in_league.csv', index=False)
team_plays_in_match.to_csv('output_csv/team_plays_in_match.csv', index=False)
```
